```python
import os
import cv2
```

```python
IMAGE_EXTENSIONS = {".jpg", ".jpeg", ".png", ".bmp"}

def collect_image_paths(dataset_dir):
    image_paths = []

    for root, _, files in os.walk(dataset_dir):
        for file in files:
            ext = os.path.splitext(file)[1].lower()

            if ext in IMAGE_EXTENSIONS:
                full_path = os.path.join(root, file)
                relative_path = os.path.relpath(full_path, dataset_dir)
                image_paths.append(relative_path)

    image_paths.sort()
    return image_paths
```

```python
def resize_to_same_height(img1, img2, target_height=480):
    def resize(img):
        h, w = img.shape[:2]
        scale = target_height / h
        new_w = int(w * scale)
        return cv2.resize(img, (new_w, target_height))

    return resize(img1), resize(img2)
```

```python
base_dir = os.getcwd()

dataset_dir = os.path.join(base_dir, "dataset/images")
gt_dir = os.path.join(base_dir, "dataset/gts")

image_paths = collect_image_paths(dataset_dir)

print(f"Found {len(image_paths)} images.")
print("Press p to proceed.")
print("Press q or Esc to quit.")

stop = False

for relative_path in image_paths:
    dataset_path = os.path.join(dataset_dir, relative_path)
    gt_path = os.path.join(gt_dir, relative_path)

    if not os.path.exists(gt_path):
        print(f"Missing GT image for: {relative_path}")
        continue

    image = cv2.imread(dataset_path)
    gt_image = cv2.imread(gt_path)

    if image is None:
        print(f"Could not read image: {dataset_path}")
        continue

    if gt_image is None:
        print(f"Could not read GT image: {gt_path}")
        continue

    image, gt_image = resize_to_same_height(image, gt_image)

    combined = cv2.hconcat([image, gt_image])

    cv2.putText(combined, "Dataset", (20, 40), cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 255, 255), 2)

    cv2.putText(combined, "Dataset_gt", (image.shape[1] + 20, 40), cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 255, 255), 2)

    cv2.imshow("Dataset vs Dataset_gt", combined)
    print(relative_path)
    if stop:
        break
    while True:
        key = cv2.waitKey(0) & 0xFF

        if key == ord(" "):
            break

        if key == ord("q") or key == 27:
            stop = True
            break

cv2.destroyAllWindows()
```
